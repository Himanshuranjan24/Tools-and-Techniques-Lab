```python
import pandas as pd
df = pd.read_csv('Salaries.csv')
Male_professors = df[df['gender'] == 'Male']
Female_professors = df[df['gender'] == 'Female']
Male_highest_salary = Male_professors.loc[Male_professors['salary'].idxmax()]
Male_lowest_salary = Male_professors.loc[Male_professors['salary'].idxmin()]
Female_highest_salary = Female_professors.loc[Female_professors['salary'].idxmax()]
Female_lowest_salary = Female_professors.loc[Female_professors['salary'].idxmin()]
print("Male Professor with Highest Salary:")
print(Male_highest_salary)
print("\nMale Professor with Lowest Salary:")
print(Male_lowest_salary)
print("\nFemale Professor with Highest Salary:")
print(Female_highest_salary)
print("\nFemale Professor with Lowest Salary:")
print(Female_lowest_salary)
```

```python
import pandas as pd
try:
  df = pd.read_csv("Salaries.csv")
  professors = df[df['rank'] == 'Prof']
  highest_paid = professors.sort_values(by='salary', ascending=False).iloc[0]
  lowest_paid = professors.sort_values(by='salary', ascending=True).iloc[0]
  print("Highest Paid Professor:")
  print(highest_paid)
  print("\nLowest Paid Professor:")
  print(lowest_paid)
except FileNotFoundError:
  print("Error: 'Salaries.csv' file not found. Please ensure the file exists and is in the correct location.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv("Salaries.csv")
male_count = data[data['gender'] == 'Male']['gender'].count()
female_count = data[data['gender'] == 'Female']['gender'].count()
labels = ['Male', 'Female']
sizes = [male_count, female_count]
colors = ['blue', 'pink']
explode = (0.1, 0)  

plt.figure(figsize=(8, 6))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
plt.title('Distribution of Male and Female Staff')
plt.axis('equal') 
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv("Salaries.csv")
prof_count = data[data['rank'] == 'Prof']['rank'].count()
assoc_prof_count = data[data['rank'] == 'AssocProf']['rank'].count()
asst_prof_count = data[data['rank'] == 'AsstProf']['rank'].count()
labels = ['Prof', 'AssocProf', 'AsstProf']
sizes = [prof_count, assoc_prof_count, asst_prof_count]
colors = ['gold', 'lightcoral', 'lightskyblue']
explode = (0.1, 0, 0)  
plt.figure(figsize=(8, 6))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
plt.title('Distribution of Professors, Associate Professors, and Assistant Professors')
plt.axis('equal') 
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv("Salaries.csv")
bin_edges = list(range(50000, int(data['salary'].max()) + 15000, 15000))
plt.figure(figsize=(10, 6))
plt.hist(data['salary'], bins=bin_edges, color='skyblue', edgecolor='black', alpha=0.7)
plt.xlabel('salary')
plt.ylabel('Frequency')
plt.title('Histogram of Salaries')
plt.xticks(bin_edges)
plt.grid(True)
plt.show()
```
